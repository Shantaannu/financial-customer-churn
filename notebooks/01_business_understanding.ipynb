## Industry Context

Telecom companies operate on a subscription based revenue model.
Customer churn directly impacts Monthly Recurring Revenue (MRR).

Key business realities:
- Acquiring a new customer is significantly more expensive than retaining an existing one (The Amount Spend on Adv, Existing Customer Knows about the product so retaining Them is Easy Rather than Spendind Amount on Advertisement For New Customer)
- Churned customers represent both revenue loss and wasted acquisition cost (Acqusition Cost Includes Marketing spend (ads, promotions, referral incentives), Sales teams efforts, Platform fees (Google Ads, affiliates, partnerships))
- Early identification of churn risk allows proactive retention strategies


## Business Problem

The telecom company is experiencing customer attrition.
Currently, churn is only observed **after** the customer has already left the service.

This reactive approach provides no opportunity for:
- Retention offers
- Service improvements
- Targeted customer engagement

The business needs a way to **predict churn in advance**.

## Business Objective

The objective of this project is to:
- Predict customers who are likely to churn
- Identify key factors contributing to churn
- Enable data-driven retention strategies

The focus is not only on prediction accuracy,
but on generating insights that the business can act upon.

## Machine Learning Problem Framing

Type of Problem: Binary Classification

Target Variable:
- Churn (Yes / No)

Model Goal:
- Estimate the probability of churn for each customer

This allows the business to prioritize high-risk customers for retention efforts.

## Why Machine Learning?

Customer churn is influenced by multiple factors such as:
- Tenure
- Contract type
- Monthly charges
- Service usage patterns

These relationships are:
- Non-linear
- Interdependent
- Not suitable for simple rule-based systems

Machine learning models can capture these complex patterns more effectively.


## Business Assumptions

The following assumptions are made for this project:
- The cost of retaining a customer is lower than acquiring a new one
- The business prefers identifying most churners, even at the cost of some false positives
- Historical customer behavior is indicative of future churn risk

## Success Metrics

Primary Metric:
- Recall for the churn class (Yes)

Reason:
- Missing a churner is more costly than incorrectly flagging a loyal customer

Secondary Metrics:
- ROC-AUC for overall model performance
- Precision to control excessive false positives

## Expected Business Impact

A successful churn prediction model will:
- Reduce customer attrition
- Improve customer lifetime value
- Enable targeted retention campaigns
- Support data-driven decision-making

This project aims to bridge the gap between analytics and business strategy.
