## Data Quality Observations

Key observations from data understanding:
- The dataset contains both numerical and categorical features
- The `TotalCharges` column is incorrectly stored as an object
- Some values in `TotalCharges` appear as empty strings
- The target variable (Churn) shows class imbalance
- No explicit timestamp features are present

These observations will guide preprocessing and feature engineering steps.


```python
df['TotalCharges'] = df['TotalCharges'].replace("",np.nan)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

df['TotalCharges'].info()
```

## TotalCharges Data Quality Fix

The `TotalCharges` column contained empty string values,
which were converted to NaN and then cast to a numeric type.

These missing values primarily correspond to customers with
very low or zero tenure, where total charges have not yet accumulated.

No imputation is performed at this stage.
The treatment of missing values will be addressed during feature engineering.
